## Matplotlib Tutorial 7 - loading data from files

```python
# Using csv
import csv


x = []
y = []

with open('Python Data/matplotlib_sentdex_sample_data_part_07.txt', 'r') as csv_file:
    plots = csv.reader(csv_file, delimiter=',')
    for row in plots:
        x.append(int(row[0]))
        y.append(int(row[1]))

plt.plot(x, y, label='Loaded from file!')

plt.legend()

plt.title('Interesting Graph\nCheck it out')
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

```python
# Using numpy
import numpy as np


# Don't have to be txt file. Can also be csv file
x, y = np.loadtxt('Python Data/matplotlib_sentdex_sample_data_part_07.txt', delimiter=',',
                  unpack=True) # unpack items, in this case, they are x and y

plt.plot(x, y, label='Loaded from file!')

plt.legend()

plt.title('Interesting Graph\nCheck it out')
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

## Matplotlib Tutorial 9 - converting data from the internet

```python
# It doesn't follow what the tutorial does.
# Used Corey Schafer's method. Check another Matplotlib jupyter notebook.
import pandas as pd


# To register the converters:
#         >>> from pandas.tseries import converter
#         >>> converter.register()
#   warnings.warn(msg, FutureWarning)
pd.plotting.register_matplotlib_converters()

def graph_data(stock):
    stock_price_url = 'https://pythonprogramming.net/yahoo_finance_replacement'
    # pd.read_csv can read from website.
    df = pd.read_csv(stock_price_url)
    df['Date'] = pd.to_datetime(df['Date'])
    
    plt.plot_date(df['Date'], df['Close'], linestyle='-', marker=',')
    
    plt.gcf().autofmt_xdate()
    
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.title('Interesting Graph\nCheck it out')
    plt.show()


graph_data('AAPL')
```

## Matplotlib Tutorial 10 - basic customizations, rotating labels

```python
# The tutorial uses subplot.
# I used plt.xticks.
def graph_data(stock):
    stock_price_url = 'https://pythonprogramming.net/yahoo_finance_replacement'
    df = pd.read_csv(stock_price_url)
    df['Date'] = pd.to_datetime(df['Date'])
    
    plt.plot_date(df['Date'], df['Close'], linestyle='-', marker=',')
    
    plt.xticks(rotation=45) # Change rotation of xticks
    
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.title('Interesting Graph\nCheck it out')
    plt.show()


graph_data('AAPL')
```

## Matplotlib Tutorial 12 - more customization of colors and fills

```python
# Change label colors and set yticks
def graph_data(stock):
    stock_price_url = 'https://pythonprogramming.net/yahoo_finance_replacement'
    df = pd.read_csv(stock_price_url)
    df['Date'] = pd.to_datetime(df['Date'])
    
    plt.plot_date(df['Date'], df['Close'], linestyle='-', marker=',')
    
    plt.gca().xaxis.label.set_color('c') # Change label color
    plt.gca().yaxis.label.set_color('r')
    plt.gca().set_yticks([0, 150, 300, 450, 600, 750, 900]) # Change yticks
    
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.title(stock)
    plt.show()


graph_data('AAPL')
```

## Matplotlib Tutorial 13 - spines and horizontal lines

```python
# Set spines
def graph_data(stock):
    stock_price_url = 'https://pythonprogramming.net/yahoo_finance_replacement'
    df = pd.read_csv(stock_price_url)
    df['Date'] = pd.to_datetime(df['Date'])
    
    plt.plot_date(df['Date'], df['Close'], linestyle='-', marker=',')
    
    # Change spines
    plt.gca().spines['left'].set_color('c')
    plt.gca().spines['right'].set_visible(False)
    plt.gca().spines['top'].set_visible(False)
    plt.gca().spines['left'].set_linewidth(5)
    
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.title(stock)
    plt.show()


graph_data('AAPL')
```

```python
# Method 1 to change the tick_params color
# The color of date was not changed. I guess it is because pd.to_datetime.
def graph_data(stock):
    stock_price_url = 'https://pythonprogramming.net/yahoo_finance_replacement'
    df = pd.read_csv(stock_price_url)
    df['Date'] = pd.to_datetime(df['Date'])
    
    plt.plot_date(df['Date'], df['Close'], linestyle='-', marker=',')
    
    plt.gca().tick_params(axis='x', color='#f06215')
    
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.title(stock)
    plt.show()


graph_data('AAPL')
```

```python
# Method 2 to change the tick_params color
def graph_data(stock):
    stock_price_url = 'https://pythonprogramming.net/yahoo_finance_replacement'
    df = pd.read_csv(stock_price_url)
    df['Date'] = pd.to_datetime(df['Date'])
    
    fig = plt.figure()
    ax = fig.add_subplot(111)
    ax.plot_date(df['Date'], df['Close'], linestyle='-', marker=',')
    
    ax.set_xlabel('Date')
    ax.set_ylabel('Price')
    ax.tick_params(axis='x', color='#f06215')
         
    plt.title(stock)
    
    plt.show()


graph_data('AAPL')
```
